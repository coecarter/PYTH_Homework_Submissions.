```python
from collections.abc import ItemsView
"""
Lists with Duplicates
"""

# First, print list a below as it is now.

a = [2, 4, 10, 20, 5, 2, 20, 4]
b = [13, 2, 25, 20, 4, 8]

print("list a:",a)
print("list b:",b)

# Using a loop, remove the duplicate items from list a and print out the updated list.

new_a = []

for num in a:
  if num not in new_a:
    new_a.append(num)
print("new_a:",new_a)

# Using a loop, merge list b into list a without adding any duplicates.

for num in b:
  if num not in new_a:
    new_a.append(num)
print("list a + list b dupes removed:",new_a)
```

```python
"""
Grades
"""
# Here's a dict containing students' grades from the semester's assignments.

import statistics as s

students = {
  'Ashton': [86, 45, 98, 100],
  'Sierra': [100, 100, 100, 100],
  'Zach': [38, 49, 90, 87],
  'Manuel': [98, 92, 86, 100],
  'Felicia': [94, 87, 89, 95],
  'Ankur': [75, 77, 77, 85],
  'Ananya': [98, 94, 87, 92],
  'Nick': [79, 84, 89, 90],
  'Olivia': [83, 91, 69, 85],
  'Molly': [83, 74, 72, 90]
}

# p 1.1) Create a dict called num_grades to store each student's average numerical grade for the semester.


num_grades = {student: s.mean(grades) for student, grades in students.items()}

print(num_grades)

# p 1.2) Create a dict called letter_grades to store each student's final letter grade for the semester. Use A, B, C, D, and F as grades per the standard grading scale.


letter_grades = {}

for student, average_grade in num_grades.items():
    if average_grade >= 90:
        letter_grades[student] = 'A'
    elif average_grade >= 80:
        letter_grades[student] = 'B'
    elif average_grade >= 70:
        letter_grades[student] = 'C'
    elif average_grade >= 60:
        letter_grades[student] = 'D'
    else:
        letter_grades[student] = 'F'

print(letter_grades)

# p 1.3) Create a list of honor roll students (i.e. students who got A's).


honor_roll = []

for student, average_grade in num_grades.items():
    if letter_grades[student] == 'A':
        honor_roll.append(student)
print('Honor Roll:',honor_roll)
```

```python
"""
Build a Triangle
"""

# Use a while loop to print a 5-level triangle of stars that looks like this:

"""
*
**
***
****
*****
"""


i = 1
while i <= 5 :
  print ("*" * i)
  i = i + 1

```

```python
"""
Build a Pyramid
"""

# Use a while loop to print a pyramid of stars that looks like this:

"""
    *
   ***
  *****
 *******
*********
"""

stars = 1
spaces = 4

while stars <= 9:
  outStr = spaces * " " + stars * "*"
  print(outStr)
  stars = stars + 2
  spaces = spaces - 1
```



### P3.PY




```python
"""
Build a Diamond
"""

# Use a while loop to print a diamond of stars that looks like this:

"""
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
"""
stars = 1
spaces = 4

# Print the top half of diamond
while stars <= 9:
    outStr = spaces * " " + stars * "*"
    print(outStr)
    stars += 2
    spaces -= 1

# Reset stars and spaces
stars = 7
spaces = 1

# Print the bottom half of diamond
while stars >= 0:
    outStr = spaces * " " + stars * "*"
    print(outStr)
    stars -= 2
    spaces += 1
```
